```python
E = np.random.random((2,3))
print('Matrix E: ]\n', E)
E = E * 10
E = np.round(E)
print('New Matrix E: ]\n', E)
print('Sum=', E.sum())
print('Max=', E.max())
print('Mim=', E.min())
print('exp(E[1])=', np.exp(E[1]))
print('sqrt(E[1])=', np.sqrt(E[1]))
```
