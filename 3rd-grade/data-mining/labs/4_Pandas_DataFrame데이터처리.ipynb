```python
import numpy as np
import pandas as pd

dico = {'X':[1,2,np.nan],'Y':[4,np.nan,np.nan],'Z':[7,8,9]}
```

```python
row_labels = 'A B C'.split()
df = pd.DataFrame(dico, row_labels)
df
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.dropna(thresh=2)
```

```python
df.fillna('xx')
```

```python
xcol_mean = df['X'].mean()
df['X'].fillna(value=xcol_mean)
```

```python
for i in 'X Y Z'.split():
  df[i].fillna(value = df[i].mean(), inplace=True)

df
```
